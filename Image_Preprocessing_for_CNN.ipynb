```python
from PIL import Image #Using Pillow library
```

**Converting RGB to Grayscale using OpenCV**

```python
# Importing OpenCV library
import cv2
```

```python
img = cv2.imread("/content/Bike.jpg")
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY) # This should be done on numpy array not on image
```

```python
print(type(grayscale_image))
print(grayscale_image.shape)
```

```python
print(grayscale_image)
```

```python
from google.colab.patches import cv2_imshow
```

```python
cv2_imshow(grayscale_image)
```

```python
# saving the grayscale using OpenCV

cv2.imwrite("Greyscale_image.jpg",grayscale_image)
```
